# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
city_data_df
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

```python
urban_cities_df = pyber_data_df[pyber_data_df["type"] == "Urban"]
urban_cities_df
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
city_type_ride_count = pyber_data_df.groupby(["type"]).count()["ride_id"]
city_type_ride_count
```

```python
# 2. Get the total drivers for each city type
city_type_driver_count = city_data_df.groupby(["type"]).sum()["driver_count"]
city_type_driver_count
```

```python
#  3. Get the total amount of fares for each city type
city_type_fare_sum = pyber_data_df.groupby(["type"]).sum()["fare"]
city_type_fare_sum
```

```python
#  4. Get the average fare per ride for each city type. 
city_type_avg_fare = pyber_data_df.groupby(["type"]).sum()["fare"]/pyber_data_df.groupby(["type"]).count()["ride_id"]
city_type_avg_fare
```

```python
# 5. Get the average fare per driver for each city type. 
city_type_avg_fare_driver = pyber_data_df.groupby(["type"]).sum()["fare"]/city_data_df.groupby(["type"]).sum()["driver_count"]
city_type_avg_fare_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame(
{"Rides by city type":city_type_ride_count,
  "Drivers by city type":city_type_driver_count,
  "Total fare by city type":city_type_fare_sum,
  "Average fare by city type":city_type_avg_fare,
  "Avg fare per driver by city type":city_type_avg_fare_driver})
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
```

```python
#  8. Format the columns.
#district_summary_df["Average Math Score"] = district_summary_df["Average Math Score"].map("{:.1f}".format)
#pyber_summary_df["Rides by city type"] = pyber_summary_df["Rides by city type"].map("{:,}".format)
pyber_summary_df["Drivers by city type"] = pyber_summary_df["Drivers by city type"].map("{:,}".format)
pyber_summary_df["Total fare by city type"] = pyber_summary_df["Total fare by city type"].map("${:,.2f}".format)
pyber_summary_df["Average fare by city type"] = pyber_summary_df["Average fare by city type"].map("${:,.2f}".format)
pyber_summary_df["Avg fare per driver by city type"] = pyber_summary_df["Avg fare per driver by city type"].map("${:,.2f}".format)
pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
fare_by_date_df = pyber_data_df.groupby(["type","date"]).sum()["fare"]
fare_by_date_df
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
fare_by_date_df = pyber_data_df.groupby(["type","date"]).sum()["fare"]
fare_by_date_df
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
fare_by_date_df = fare_by_date_df.reset_index()
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
fare_by_date_df.pivot(index='date',columns='type',values='fare')
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
fare_jan_to_april_df = fare_by_date_df.loc[(fare_by_date_df["date"]>="2019-01-01 00:00:00")&(fare_by_date_df["date"]<="2019-04-29 23:59:59")]
fare_jan_to_april_df
```

```python
fare_jan_to_april_df.info()
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
fare_jan_to_april_df.index = pd.to_datetime(fare_jan_to_april_df.date)
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
fare_jan_to_april_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
#fare_jan_to_april_df["type"].astype('float')
fare_by_week_df = fare_jan_to_april_df.groupby("type").resample('W').sum()
fare_by_week_df =  fare_by_week_df.reset_index()
#fare_by_week_df2 = fare_by_week_df.groupby("type")
fare_by_week_df=fare_by_week_df.pivot(index='date',columns='type',values='fare')
fare_by_week_df=fare_by_week_df.iloc[0:17]
```

```python
fare_by_week_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 
#fig, ax = plt.subplots()
fare_by_week_df.plot(figsize=(20,6))

plt.title("Total fare by city type")
plt.ylabel("Fare ($USD)")
#plt.xlabel("Total Number of Rides (Per City)")
# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
plt.savefig("analysis/pyber_fare_summary.png")

plt.show()
```
